```python
import os
import shutil

def extract_pdfs_from_directory(root_directory, destination_folder):
    for subdir, dirs, files in os.walk(root_directory):
        for file in files:
            if file.endswith(".pdf"):
                pdf_file_path = os.path.join(subdir, file)
                shutil.copy(pdf_file_path, destination_folder)
```
