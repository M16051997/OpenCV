#### Bitwise Operations

    The cv2.bitwise_and() function takes two input arrays and performs the bitwise AND operation on them. It is useful for tasks such as image masking, feature extraction, and bit-level operations on images. Make sure that the input arrays have the same size and type, and the result will also be of the same size and type.

```python
import cv2
import numpy as np

# Create two sample arrays
array1 = np.array([1, 1, 0, 0], dtype=np.uint8)
array2 = np.array([1, 0, 1, 0], dtype=np.uint8)

# Perform bitwise AND operation
result_array = cv2.bitwise_and(array1, array2)

# Print the result
print("Result Array:", result_array)
```

```python
import cv2 as cv

blank = np.zeros((400,400), dtype='uint8')

rectangle = cv.rectangle(blank.copy(), (30,30), (370,370), 255, -1)
circle = cv.circle(blank.copy(), (200,200), 200, 255, -1)

cv.imshow('rectangle',rectangle)
cv.imshow('Circle', circle)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# bitwise AND --> intersecting regions

bitwise_and = cv.bitwise_and(rectangle, circle)
cv.imshow('Bitwise and', bitwise_and)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# bitwise OR --> non-intersecting and intersecting regions
bitwise_or = cv.bitwise_or(rectangle, circle)
cv.imshow('Bitwise OR', bitwise_or)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# bitwise XOR --> non-intersecting regions
bitwise_xor = cv.bitwise_xor(rectangle, circle)
cv.imshow('Bitwise XOR', bitwise_xor)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# bitwise NOT
bitwise_not = cv.bitwise_not(circle)
cv.imshow('Circle NOT', bitwise_not)
cv.waitKey(0)
cv.destroyAllWindows()
```

### Gradients

```python
### Sobel Operator

import cv2 as cv
import numpy as np

img = cv.imread(r"C:\Users\50510\Desktop\Cat.jpg")

# Apply the Sobel operator
sobelx = cv.Sobel(img, cv.CV_64F, 1, 0, ksize=5)
sobely = cv.Sobel(img, cv.CV_64F, 0, 1, ksize=5)

# Combine the results
sobel_comined = cv.addWeighted(sobelx, 0.5, sobely, 0.5, 0)

cv.imshow('Sobel X', sobelx)
cv.imshow('Sobel Y', sobely)
cv.imshow('Combined Sobel', sobel_comined)

cv.waitKey(0)
cv.destroyAllWindows()
```

```python
### Sobel Operator

import cv2 as cv
import numpy as np

img = cv.imread(r"C:\Users\50510\Desktop\Cat.jpg")
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

# Apply the Sobel operator
sobelx = cv.Sobel(gray, cv.CV_64F, 1, 0, ksize=5)
sobely = cv.Sobel(gray, cv.CV_64F, 0, 1, ksize=5)

# Combine the results
sobel_comined = cv.addWeighted(sobelx, 0.5, sobely, 0.5, 0)

cv.imshow('Sobel X', sobelx)
cv.imshow('Sobel Y', sobely)
cv.imshow('Combined Sobel', sobel_comined)

cv.waitKey(0)
cv.destroyAllWindows()
```

```python
def is_gray_scale(image_path):
    img = cv2.imread(image_path, cv2.IMREAD_UNCHANGED)

    if len(img.shape) < 3:  # Check if the image has only 2 dimensions
        return True
    elif len(img.shape) == 3:  # Check if the image has 3 dimensions
        height, width, channels = img.shape
        if channels == 1:  # Check if the image has 1 channel
            return True
        elif channels == 3:  # Check if the image has 3 channels
            b, g, r = img[:, :, 0], img[:, :, 1], img[:, :, 2]
            if (b == g).all() and (b == r).all():  # Check if all channels are equal
                return True
            else:
                return False
        else:
            return False
    else:
        return False
    
# Provide the path to the image file
image_path = 'path_to_your_image_file'

# Call the function and print the result
if is_gray_scale(image_path):
    print("The image is grayscale.")
else:
    print("The image is colored.")
```
